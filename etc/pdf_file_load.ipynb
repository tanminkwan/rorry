```python
import fitz  # PyMuPDF

def pdf_page_iterator(pdf_path):
    # PDF 파일 열기
    pdf_document = fitz.open(pdf_path)
    
    # 각 페이지를 순회하며 텍스트를 yield
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        page_text = page.get_text()
        yield page_num + 1, page_text

# 예시 PDF 파일 경로
pdf_path = '2021_korea_bank.pdf'

# PDF의 각 페이지를 이터레이터로 처리
for page_num, page_text in pdf_page_iterator(pdf_path):
    print(f"Page {page_num}:\n{page_text}\n")
    if page_num > 30:
        break
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# PDF 파일 경로
file_path = '2021_korea_bank.pdf'

# PDF 파일 로더 초기화
loader = PyPDFLoader(file_path)

# PDF 파일을 로드하여 전체 텍스트 추출
pages = loader.load()

# 전체 텍스트를 하나의 문자열로 결합
full_text = ''.join(page.page_content for page in pages)

# 텍스트 스플리터 설정
chunk_size = 1000
chunk_overlap = 100
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunk_size, 
    chunk_overlap=chunk_overlap
)

# 전체 텍스트를 나누기
chunks = text_splitter.split_text(full_text)

# 각 덩어리의 내용을 출력
for i, chunk in enumerate(chunks):
    print(f"Chunk {i + 1}:\n{chunk}\n")
    if i > 30:
        break
```
